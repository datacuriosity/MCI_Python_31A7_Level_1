```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
iris.head(2)
```

```python
iris.species.value_counts()
```

```python
iris.groupby('species').species.count()
```

#### Scatter plot

```python
plt.scatter(iris.sepal_length, iris.sepal_width)
plt.title('Sepal length vs Sepal Width')
plt.xlabel('Sepal length')
plt.ylabel('Sepal witdth');
```

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width');
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width');
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species');
```

```python
ax = plt.axes()
iris.query("species == 'setosa'").plot(kind='scatter', x='sepal_length', y='sepal_width', label='setosa', ax=ax)
iris.query("species == 'versicolor'").plot(kind='scatter', x='sepal_length', y='sepal_width', label='versicolor', 
                                           ax=ax, color='orange')
iris.query("species == 'virginica'").plot(kind='scatter', x='sepal_length', y='sepal_width', label='virginica', 
                                          ax=ax, color='red')
plt.legend(loc='upper right');
```

### Histogram

```python
sns.displot(data=iris, x='sepal_length', kind='hist'); 
#kde --> kernel density estimation
```

```python
sns.displot(data=iris, x='sepal_length', kind='kde');  #kde --> kernel density estimation
```

### Multiple plot

```python
sns.pairplot(data=iris, diag_kind='kde');
```

### Mix matplotlib vs seaborn

```python
_, ax = plt.subplots(1, 2, figsize=(15, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', ax=ax[0])
ax[1].hist(iris.petal_length, bins=50);
```

- Numeric:

- Categorical: 
    - Ordinal: ranked --> Excellent, Good, Medium, Bad. 
    - Nominal: Gender, Smoker,..

### Facet Grid

```python
tips.head(2)
```

```python
g = sns.FacetGrid(data=tips, col='sex')
g.map(sns.scatterplot, 'total_bill', 'tip');
```

```python
g = sns.FacetGrid(data=tips, col='sex', row='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip');
```

```python
g = sns.FacetGrid(data=tips, col='sex', row='smoker', hue='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.legend();
```

### Categorical Plot

```python
# aspect = width / height
sns.catplot(data=tips, x='time', y='total_bill', kind='strip', height=4, aspect=1.5, col='sex'); 
```

```python
# aspect = width / height
sns.catplot(data=tips, x='time', y='total_bill', kind='violin'); 
```

```python
# aspect = width / height
sns.catplot(data=tips, x='time', y='total_bill', kind='box'); 
```
