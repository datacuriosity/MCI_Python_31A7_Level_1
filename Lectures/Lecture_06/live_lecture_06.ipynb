```python
x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Sin function')
plt.plot(x, np.cos(x), label='Cosin function')
plt.title("Sin function")
plt.xlabel('Linear space data')
plt.ylabel('Sin values')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x) + np.random.uniform(size=100)
z = np.cos(x) + np.random.uniform(size=100)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, label='Sin function')
plt.scatter(x,z, label='Cosin function')
plt.title("Sin function")
plt.xlabel('Linear space data')
plt.ylabel('Sin values')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

- Mean, Median, Mode --> Central Tendency

```python
# Standard Distribution --> mean = 0, std = 1
x = np.random.normal(size=100000)
plt.hist(x, bins=50);
```

```python
# Standard Distribution --> mean = 0, std = 1
x = np.random.normal(loc=2, scale=2, size=100000)
plt.hist(x, bins=50);
```

```python
x = np.random.uniform(size=100000)
plt.hist(x, bins=20);
```

```python
x = np.random.normal(loc=2, scale=2, size=100000)
plt.boxplot(x);
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
tips.time.value_counts().plot(kind='bar');
```

```python
plt.bar(tips.time.value_counts().index, height=tips.time.value_counts());
```

```python
tips.pivot_table(index='day', columns='smoker', values='total_bill', aggfunc='sum').plot(kind='bar');
```

```python
tips.head(2)
```

```python
plt.pie(tips.groupby(['day']).total_bill.sum(), labels=tips.groupby(['day']).total_bill.sum().index, 
        autopct='%.2f%%', explode=[0, 0.2, 0, 0]);
```

```python
tips.groupby(['day']).total_bill.sum().plot(kind='pie', autopct='%.2f%%', explode=[0, 0.2, 0, 0]);
```

```python
tips.head(2)
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.scatter(tips.total_bill, tips.tip, color='purple')
plt.title('Total bill vs tip')

plt.subplot(1, 2, 2)
tips.groupby('sex').total_bill.sum().plot.bar(color='red', title='Total bill for each gender')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 6))
ax[0].scatter(tips.total_bill, tips.tip, color='purple')
ax[0].set(title='Total bill vs tip', xlabel='Total bill', ylabel='tip')
tips.groupby('sex').total_bill.sum().plot.bar(color='red', title='Total bill for each gender', ax=ax[1], ylabel='Total bill');
```
